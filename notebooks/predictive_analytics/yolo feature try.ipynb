```python
df_bike_trips = pd.read_parquet('../../data/bike_trips_cleaned.parquet')
df_bike_trips.head()
```

```python
## next few cells need to be moved into the bike_trips notebook at some point because we want to have readable datetimes everywhere 
## not just in this notebook

df_bike_trips['start_time'] = pd.to_datetime(df_bike_trips['start_time'])
df_bike_trips['end_time'] = pd.to_datetime(df_bike_trips['end_time'])
```

```python
datetime_format = '%d.%m.%Y %H:%M:%S'
print(f"Earliest start observation: {format(df_bike_trips['start_time'].min(), datetime_format)}")
print(f"Latest start observation: {format(df_bike_trips['start_time'].max(), datetime_format)}")

print(f"Earliest end observation: {format(df_bike_trips['end_time'].min(), datetime_format)}")
print(f"Latest end observation: {format(df_bike_trips['end_time'].max(), datetime_format)}")
```

```python
df_bike_trips = df_bike_trips[
    (df_bike_trips["start_time"] >= "2016-01-01 00:00:00")
    & (df_bike_trips["start_time"] <= "2016-12-31 23:59:59")
    ]

df_bike_trips = df_bike_trips[
    (df_bike_trips["end_time"] >= "2016-01-01 00:00:00")
    & (df_bike_trips["end_time"] <= "2016-12-31 23:59:59")
    ]
```
